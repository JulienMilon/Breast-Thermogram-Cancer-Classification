```python
# create figure 
fig = plt.figure(figsize=(10, 7)) 
  
# setting values to rows and column variables 
rows = 2
columns = 2

# Adds a subplot at the 1st position 
fig.add_subplot(rows, columns, 1) 

# showing image 
plt.imshow(image) 
plt.axis('off') 
plt.title("No Processing") 
  
# Adds a subplot at the 2nd position 
fig.add_subplot(rows, columns, 2) 
  
# showing image 
plt.imshow(equ) 
plt.axis('off') 
plt.title("Histograms Equalization") 
  
# Adds a subplot at the 3rd position 
fig.add_subplot(rows, columns, 3) 
  
# showing image 
plt.imshow(heatmap_turbo) 
plt.axis('off') 
plt.title("Heatmap No processing") 
  
# Adds a subplot at the 4th position 
fig.add_subplot(rows, columns, 4) 
  
# showing image 
plt.imshow(heatmap_equ) 
plt.axis('off') 
plt.title("Heatmap Histograms Equalization") 
```

```python
input_path = "../../../DataSet/DMR - Database For Mastology Research - Visual Lab/Healthy/"
output_path = "../../../DataSet/Transformed - DMR - Database For Mastology Research - Visual Lab/Healthy/"


for name in os.listdir(input_path):
    #print(name)
    image = cv2.imread(input_path + name, 0)
    equ = cv2.equalizeHist(image) # Histograms Equalization
    heatmap_equ = cv2.applyColorMap(equ, cv2.COLORMAP_TURBO) # Processing from Gray IR Image to RGB.
    cv2.imwrite(output_path + name, heatmap_equ)
```
